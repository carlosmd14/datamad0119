```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline
from os import path
from PIL import Image
```

```python
df1 = pd.read_csv('df_sin_dummies')
```

```python
df1.genres[0]
```

```python
genres = df1.genres.tolist()
```

```python
flat_gen = [item for sublist in genres for item in sublist]

```

```python
genres
```

```python
genres_str = ''.join(flat_gen)
```

```python
genres_str = genres_str.replace("'", "")
genres_str = genres_str.replace("[", "")
genres_str = genres_str.replace("]", ", ")
genres_str
```

```python
text = genres_str

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
