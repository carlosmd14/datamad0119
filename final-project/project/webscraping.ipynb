#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
html = requests.get(url).content
soup = BeautifulSoup(html)
````

````python
#your code

link_tags = soup.findAll('a')

link_urls = []

for link_tag in link_tags:
    if link_tag.get('href') is None:
        pass
    elif link_tag.get('href').startswith('/wiki'):
        link_urls.append('https://en.wikipedia.org' + link_tag.get('href'))
    else:
        link_urls.append('https://en.wikipedia.org/wiki/Python_(programming_language)' + link_tag.get('href'))
    
link_urls
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
html = requests.get(url).content
soup = BeautifulSoup(html)
````

````python
#your code

table = soup.find('table', attrs={'border':'0'})


table_rows = table.find_all('tr')
l = []
for tr in table_rows:
    td = tr.find_all('td')
    if tr.text.startswith('earthquake'):
        row = [re.sub('..min ago$|.hr ..min ago$', '', tr.text.replace("\xa0", "").lstrip("earthquake")) for tr in td]
        l.append(row)
l
df = pd.DataFrame(l, columns=['1','2','3','Date&Time','Latitude','4','Longitude','5','6','7','8','Region_Name','Update'])
df = df.drop(columns=[str(x) for x in range(1,9)])
df
````
