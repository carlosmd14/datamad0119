```python
import json
import requests
from bs4 import BeautifulSoup
import pandas as pd
from html2text import unescape
```

```python
def q_list(n=1):  
    ''' Esta función devuelve una lista de 50*n preguntas de trivia, pudiendo estar repetidas como máximo 
    n veces cada pregunta, ya que la API devuelve un máximo de 50 preguntas aleatorias por solicitud
    '''
    ql = [requests.get('https://opentdb.com/api.php?amount=50').json()['results'] for i in range(n)]
    return [item for sublist in ql for item in sublist]
```

```python
trivia_df = pd.DataFrame(q_list(24))
trivia_df
```

```python
trivia_df['incorrect_answers'] = trivia_df['incorrect_answers'].apply(lambda x: ','.join(map(str, x)))
trivia_df.tail()
```

```python
for column in trivia_df:
    trivia_df[column] = trivia_df[column].apply(unescape)

trivia_df.head()
```
