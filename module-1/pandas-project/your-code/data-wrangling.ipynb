```python
# Paso 0

import pandas as pd
import numpy as np
```

```python
# Paso 1

data = pd.read_csv('./Sharks2.csv')
data
```

```python
# Paso 2

before = len(data)
data = data.drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
# Paso 3

drop_cols1 = ['Case Number', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22', 'Unnamed: 23']
data = data.drop(drop_cols1, axis=1)
data
```

```python
# Paso 4

null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

```python
# Paso 5

data = data.drop(['Date', 'Age', 'Time'], axis=1)
data
```

```python
# Paso 6

stats = data.describe(include='all')
stats
```

```python
# Paso 7

data.dtypes
```

```python
# Paso 8

data["Country"]=data["Country"].fillna("")
data["Area"]=data["Area"].fillna("")
```

```python
# Paso 9

np.unique(data["Country"])
```

```python
# Paso 10

data['Country'] = data['Country'].str.replace('?', '')
data['Country'] = data['Country'].str.replace('&', 'AND')
data['Country'] = data["Country"].str.upper()
```

```python
# Paso 11

data['Country'].value_counts()
```

```python
# Paso 12

data['Fatal (Y/N)'].value_counts()
```

```python
# Paso 13

data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace(' ', '')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.upper()
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('#VALUE!', 'UNKNOWN')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('F', 'UNKNOWN')
```

```python
# Paso 14

data['Fatal (Y/N)'].value_counts(normalize=True)*100
```

```python
# Paso 15

type_counts = data['Type'].value_counts(normalize=True)*100
type_counts
```

```python
# Paso 16

year_labels = ['Other', '1950', '1960', '1970', '1980', '1990', '2000', '2010']
cutoffs_year = [-1, 1949, 1959, 1969, 1979, 1989, 1999, 2009, 2019]
bins_year = pd.cut(data['Year'], cutoffs_year, labels=year_labels)
bins_year
```

```python
# Paso 17

bins_year.value_counts(sort=False, normalize=True)*100
```

```python
# Paso 18

data['Country'].value_counts(normalize=True)*100
```

```python
# Paso 19

data.groupby(['Country', 'Fatal (Y/N)']).size().sort_values(ascending=False)
```

```python
# Paso 20

data.to_csv('./Sharks_After.csv', index=False)
```
